## Labelling

```python
def check_early_delinquency(loan_age, payment_history):
    # Checking current loan age. We will check early delinquency status if the loan age is under 12.
    if loan_age >= 0 and loan_age <= 12:
        # Deciding range according to loan age
        start_pos = (-2 * loan_age) - 2
        first_months_history = payment_history[start_pos:]


        for i in range(0, len(first_months_history), 2):
            month_status = first_months_history[i:i+2]
            if month_status not in ('00', 'XX'):
                return 1
    return 0


"""
The format of Delinquency status is below.
XX...XX (total 48) -> 0 month (start month)
XX...XXOO -> 1st month
XX...XXOOOO -> 2nd month.
So, because we want to chech early delinquency at first 12 months, we divide the right 24 characters from whole string.
And I detected if there were any strings other than OO or XX (ex: 01, 02) in that string.
"""
```

```python
# Changing the type of variable
df['Loan Age'] = df['Loan Age'].astype(int)
```

```python
df['Delinquency(1 or 0)'] = df.apply(lambda row: check_early_delinquency(row['Loan Age'], row['Loan Payment History']), axis=1)

early_delinquency_df = df[df['Delinquency(1 or 0)'] == 1]
print(early_delinquency_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, brier_score_loss, classification_report
import scikitplot as skplt
import matplotlib.pyplot as plt
from hmeasure import h_score
import pickle
```

```python
data = pd.read_csv('/content/drive/MyDrive/Predicting_Early_Delinquency/Dataset_for_Modeling')

features = ['Original interest Rate', 'Original Loan Term', 'Origination Date', 'Original Combined Loan to Value Ratio',
            'Number of Borrowers', 'Debt-To-Income', 'Borrower Credit Score at Origination',
            'First Time Home Buyer Indicator', 'Loan Purpose', 'Proper Type', 'Numbers of Units',
            'Occupancy Status', 'SATO', 'HPIs', 'Unemployment Rates', 'BAA Spread']
label = 'Delinquency(1 or 0)'

# Re-formatting data type.
data['Original interest Rate'] = data['Original interest Rate'].astype(float)
data['Original Loan Term'] = data['Original Loan Term'].astype(int)
data['Origination Date'] = data['Origination Date'].astype(str)
data['Original Combined Loan to Value Ratio'] = data['Original Combined Loan to Value Ratio'].astype(float)
data['Number of Borrowers'] = data['Number of Borrowers'].astype(int)
data['Debt-To-Income'] = data['Debt-To-Income'].astype(float)
data['Borrower Credit Score at Origination'] = data['Borrower Credit Score at Origination'].astype(float)
data['First Time Home Buyer Indicator'] = data['First Time Home Buyer Indicator'].apply(lambda x: 1 if x == 'Y' else 0)
data['Loan Purpose'] = data['Loan Purpose'].astype('category')
data['Proper Type'] = data['Proper Type'].astype('category')
data['Numbers of Units'] = data['Numbers of Units'].astype('category')
data['Occupancy Status'] = data['Occupancy Status'].astype('category')
data['SATO'] = data['SATO'].astype(float)
data['HPIs'] = data['HPIs'].astype(float)
data['Unemployment Rates'] = data['Unemployment Rates'].astype(float)
data['BAA Spread'] = data['BAA Spread'].astype(float)
```

```python
# If there is Nan value in such row, drop that row.
data.dropna(subset=features + [label], inplace=True)


# Defining numberical feature and categoricala feature.
numeric_features = ['Original interest Rate', 'Original Loan Term', 'Original Combined Loan to Value Ratio',
                    'Number of Borrowers', 'Debt-To-Income', 'Borrower Credit Score at Origination',
                    'SATO', 'HPIs', 'Unemployment Rates', 'BAA Spread']
categorical_features = ['Origination Date', 'First Time Home Buyer Indicator', 'Loan Purpose', 'Proper Type', 'Numbers of Units', 'Occupancy Status']

# Defining pre-processing pipeline
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])


# Dividing bad data and good data
bad_data = data[data[label] == 1]
good_data = data[data[label] == 0]
```

## Normal Modeling - Logistic Regression, Randomforest

```python
np.random.seed(42)
good_sample = good_data.sample(frac=0.01, random_state=42)

sampled_data = pd.concat([bad_data, good_sample])

X_sampled = sampled_data[features]
y_sampled = sampled_data[label]

X_train, X_test, y_train, y_test = train_test_split(X_sampled, y_sampled, test_size=0.2, random_state=42)

logreg_sampled = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('classifier', LogisticRegression(max_iter=1000))])

rf_sampled = Pipeline(steps=[('preprocessor', preprocessor),
                             ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))])

sampled_models = {'Logistic Regression': logreg_sampled, 'Random Forest': rf_sampled}
sampled_results = {}

for name, model in sampled_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    y_test_np = np.array(y_test)  # Convert y_test to numpy array
    y_prob_np = np.array(y_prob)


    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    brier = brier_score_loss(y_test, y_prob)
    h_measure = h_score(y_test_np, y_prob_np)
    report = classification_report(y_test, y_pred)
    sampled_results[name] = {'accuracy': accuracy, 'AUC': auc, 'Brier Score': brier, 'H Measure': h_measure, 'classification_report': report}

    print(f"Model: {name}")
    print(f"Accuracy: {accuracy}")
    print(f"AUC: {auc}")
    print(f"Brier Score: {brier}")
    print(f"H Measure: {h_measure}")
    print(f"Classification Report:\n{report}")

    skplt.metrics.plot_roc(y_test, model.predict_proba(X_test))
    plt.title(f'ROC Curve for {name}')
    plt.show()

    skplt.metrics.plot_precision_recall(y_test, model.predict_proba(X_test))
    plt.title(f'Precision-Recall Curve for {name}')
    plt.show()

    skplt.metrics.plot_calibration_curve(y_test, [y_prob])
    plt.title(f'Calibration Curve for {name}')
    plt.show()
```

## Using Iteration method to supplement sampling method's limitation.

```python
n_iterations = 10
results = {'Logistic Regression': [], 'Random Forest': []}

for i in range(n_iterations):
    good_sample = good_data.sample(frac=0.01, random_state=i)

    sampled_data = pd.concat([bad_data, good_sample])


    X_sampled = sampled_data[features]
    y_sampled = sampled_data[label]

    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(X_sampled, y_sampled, test_size=0.2, random_state=i)

    # Logistic Regression
    logreg_sampled = Pipeline(steps=[('preprocessor', preprocessor),
                                     ('classifier', LogisticRegression(max_iter=1000))])

    # Random Forest
    rf_sampled = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('classifier', RandomForestClassifier(n_estimators=100, random_state=i))])

    sampled_models = {'Logistic Regression': logreg_sampled, 'Random Forest': rf_sampled}

    for name, model in sampled_models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        y_test_np = np.array(y_test)
        y_prob_np = np.array(y_prob)

        accuracy = accuracy_score(y_test_np, y_pred)
        auc = roc_auc_score(y_test_np, y_prob_np)
        brier = brier_score_loss(y_test_np, y_prob_np)
        h_measure = h_score(y_test_np, y_prob_np)

        results[name].append({'accuracy': accuracy, 'AUC': auc, 'Brier Score': brier, 'H Measure': h_measure})

        model_filename = f'/content/drive/MyDrive/Predicting_Early_Delinquency/{name.lower().replace(" ", "_")}_model_iteration_{i}.pkl'
        with open(model_filename, 'wb') as file:
            pickle.dump(model, file)


for name, metrics in results.items():
    avg_accuracy = np.mean([metric['accuracy'] for metric in metrics])
    avg_auc = np.mean([metric['AUC'] for metric in metrics])
    avg_brier = np.mean([metric['Brier Score'] for metric in metrics])
    avg_h_measure = np.mean([metric['H Measure'] for metric in metrics])

    print(f"{name} Model's Mean Performance:")
    print(f"Accuracy: {avg_accuracy}")
    print(f"AUC: {avg_auc}")
    print(f"Brier Score: {avg_brier}")
    print(f"H Measure: {avg_h_measure}")
    print("\n")
```

## Using Voting

```python
from sklearn.ensemble import VotingClassifier


n_estimators = 10
estimators = []

for i in range(n_estimators):
    good_sample = good_data.sample(frac=0.01, random_state=i)
    sampled_data = pd.concat([bad_data, good_sample])
    X_sampled = sampled_data[features]
    y_sampled = sampled_data[label]
    X_train, X_test, y_train, y_test = train_test_split(X_sampled, y_sampled, test_size=0.2, random_state=i)

    rf_model = Pipeline(steps=[('preprocessor', preprocessor),
                               ('classifier', RandomForestClassifier(n_estimators=100, random_state=i))])

    rf_model.fit(X_train, y_train)
    estimators.append((f'rf_{i}', rf_model))


ensemble_model = VotingClassifier(estimators=estimators, voting='soft')
ensemble_model.fit(X_train, y_train)

# Saving model
ensemble_model_filename = '/content/drive/MyDrive/Predicting_Early_Delinquency/ensemble_random_forest_model.pkl'
with open(ensemble_model_filename, 'wb') as file:
    pickle.dump(ensemble_model, file)


y_pred = ensemble_model.predict(X_test)
y_prob = ensemble_model.predict_proba(X_test)[:, 1]

y_test_np = np.array(y_test)
y_prob_np = np.array(y_prob)

accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)
brier = brier_score_loss(y_test, y_prob)
h_measure = h_score(y_test_np, y_prob_np)

print(f"Ensemble Model's Performance:")
print(f"Accuracy: {accuracy}")
print(f"AUC: {auc}")
print(f"Brier Score: {brier}")
print(f"H Measure: {h_measure}")
```
